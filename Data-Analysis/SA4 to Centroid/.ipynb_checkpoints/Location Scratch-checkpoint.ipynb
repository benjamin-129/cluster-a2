```python
# loc_data = json.load(open("sa2_geojson_low.json"))
loc_data = json.load(open("sa4_geojson.json"))
```

```python
# loc_data['features'][0]
```

```python
loc_data['features']
```

```python
sa4_processed = {}

for sa4 in loc_data['features']:
    
    code = sa4['properties']['SA4_CODE16']
    try:
        coord = sa4['geometry']['coordinates']

        x = coord[1]
        y = coord[0]
        processed_coord = [x, y]
        
        area = sa4['properties']['AREASQKM16']

        sa4_processed[code] = {'coord':coord, 'processed_coord':processed_coord, 'area':area}
    
    except TypeError:
        continue

```
