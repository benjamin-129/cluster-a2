```python
sa4_processed = {}

for sa4 in loc_data['features']:
    
    code = sa4['properties']['SA4_CODE16']
    try:
        coord = sa4['geometry']['coordinates']

        x = coord[1]
        y = coord[0]
        processed_coord = [x, y]

        sa4_processed[code] = {'coord':coord, 'processed_coord':processed_coord}
    
    except TypeError:
        continue

```

```python
sa4_processed = {}
for sa4 in loc_data['features']:
    try:
        a= sa4['properties']['SA4_CODE16']
        b= sa4['geometry']['coordinates']
        
        sa4_processed[a]=b
    except TypeError:
        continue
```
